Import sympy, initialize pretty printing for it, and enable interactive plotting widgets.


```python
import sympy
sympy.init_printing()
%matplotlib widget
```

We can create variables, which are then displayed in a nice latex format. Don't `print` them, though: That will not be pretty.

```python
x,y = sympy.symbols('x y')
x, y
```

We can define functions of `x` using the symbol we defined above.

```python
polynomialFunc = (2*x + 3)**4
polynomialFunc
```

```python
polynomialFunc.diff(x)
```

```python
sympy.plot(polynomialFunc)
```

```python
sympy.plot_implicit(sympy.LessThan(sympy.Abs(x - y), 1))
```

```python
sympy.plot_implicit(sympy.Eq(sympy.Abs(x) + sympy.Abs(y), 1))
```

```python
sympy.plot_implicit(sympy.Eq(sympy.Abs(x) - sympy.Abs(y), 1))
```

```python
sympy.plot_implicit(sympy.Eq(sympy.Abs(x-1), sympy.Abs(y-1)))
```

```python
f_one = x / (1 + x**2)
sympy.plot(f_one)
```

```python
f_two = x*sympy.sin(1/x)
sympy.plot(f_two)
```

```python
f_two = x**2 * sympy.sin(1/x)
sympy.plot(f_two)
```
